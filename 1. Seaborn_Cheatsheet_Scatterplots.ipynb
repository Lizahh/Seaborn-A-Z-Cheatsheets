# **Part A. Simplest Plot - Scatter Plot**


* Scatterplot shows the **relationship between two continous features.**

* **Continous features** are numerical variables that can take any number of values between 2 values e.g. price = 2500. **whereas categorical faetures** has string values e.g. color = 'red'.



```python
import seaborn as sns
import pandas as pd

df = pd.read_csv('/content/sample_data/dm_office_sales.csv')

df.head()
```

```python
sns.scatterplot(x = 'salary', y = 'sales', data = df)
```

## **Parameters in Scatterplot:**

**1. 'hue' parameter:**

  * It is used to differentiate data points in scatterplot based on the value of this parameter.
  * It takes in a column as input. Always use categorical column (column that has string values) as hue parameter. Seaborn doesn't allow to use continous column as hue BUT! only basic plots like Scatterplot and lineplot allow us to use continous column too, as hue but THAT IS NOT RECOMMENDED!
  * So by adding hue, we can see 3-dimensional information on 2-dim plot :)



```python
sns.scatterplot(x = 'salary', y = 'sales', data = df, hue = 'level of education');
```

In hue, we can use these attributes to enhance its usage:

  * We can change the **palette** of the colors used in hue. There are 6 palette available: deep, muted, pastel, bright, dark, colorblind.

  * We can change the size of hue by giving a constant size e.g. 200 or we can also use some column as size of hue using **size or s** keyword


```python
# using palette and size dependent on any column's data values

sns.scatterplot(x = 'salary', y = 'sales', data = df, hue = 'level of education', palette = 'pastel', size = 'salary');
```

```python
# using palette and constant size s

sns.scatterplot(x = 'salary', y = 'sales', data = df, hue = 'level of education', palette = 'pastel', s = 200);
```

**2.  Alpha Parameter:**

We can use **alpha** parameter to control opacity/transparency of the plotted elements (0 = totally invisible and 1 = totally visible).

**3. Style Parameter:**

  * We can change the style of markers/data points in scatterplot using **style** parameter. Use any column name in it because on basis of the values of that column, it will make the marker for that datapoint.

```python
# using alpha and style parameter

sns.scatterplot(x = 'salary', y = 'sales', data = df, hue = 'level of education', alpha = 0.5, style = 'training level');
```

So, we can say:

Only hue = different colors
Only style = different shapes

hue + style = different colored shapes

**Nota Bene:**

**Style and hue both should be categorical columns**. We can use same column as hue and style too. 
But style must be a categorical column even in scatterplot, because we are categorizing items into different shapes/markers.
