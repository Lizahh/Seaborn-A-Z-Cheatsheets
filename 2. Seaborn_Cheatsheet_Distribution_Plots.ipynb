# **Introduction:**

1. Distribution plots display a single continous feature or relationship between 2 continous features.

2. They help visualize properties such as derivation and average values etc.

3. There are 3 main distribution plot types:

  * Rug Plot
  * Histogram
  * KDE Plot (Kernel Density Estimation)

# **1. Rug Plot:**

1. It is simplest distribution plot.

2. It only adds a dash/tick line for every datapoint.

3. The y-axis in it doesn't has any meaning. 

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('/content/sample_data/dm_office_sales.csv')

sns.rugplot(x = 'salary', data = df);
```

Change the height of these ticks using **height** parameter.

Give its value between 0-1.
0.5 means make it heighted till 50% height of canvas. 

```python
sns.rugplot(x = 'salary', data = df, height = 0.6); 
```

**Drawback:**

Can you count how many ticks are there at 60000? Can't tell. Hmm :/

Its solution is to use Histogram or displot.



# **2. Histplot:**

* Histograms are typically used in seaborn instead of rug plots when you want to **visualize the overall distribution and frequency of data points.**

* **Just decide number of bins in it.** (Lemme tell u in simple words: 3 bins means 3 rectangles (bars), 10 bins means 10 rectangles :P) **Bins basically tells in how many chunks you are gonna cut to x-axis.** 

* Increasing number of bins show more detailed view instead of general trends.

* So if you wanna know how many datapoints are there at 60000, you can tell by looking at the size of bar at 60000. Solution of rugplot drawback :))

* In histplot, y-axis shows the count or percentage. By percentage we mean, that if its on 0.5, it means 50% values of all the points in this bar are in 10-20 etc.





```python
sns.histplot(x = 'salary', data = df)
```

```python
# use bins now

sns.histplot(x = 'salary', data = df, bins = 4);
```

* We can add **background** to this histplot using **sns.set(style=)** parameter. Give it values as 'white' (by deault), 'dark', 'ticks'.

* Change the color of bars using **color** parameter.

* Change the color of edges of bars using **edgecolor** parameter.

* Change the width of edge lining using **linewidth** parameter.

* Change the style of the parameter using **ls** parameter.

```python
sns.set(style = 'dark')
sns.histplot(x = 'salary', data = df, color = 'red', edgecolor = 'purple', linewidth = 3, ls = '-.');
```

```python
# Changing style to ticks
sns.set(style = 'ticks')
sns.histplot(x = 'salary', data = df, color = 'red', edgecolor = 'purple');
```

# **3. KDE (Kernel Density Estimation) Plot:**

1. It is a plotting curve. It is a method of estimating probability density function of a **continous variable.**


2. In simple words, it is continous probability curve for a finite data sample.

3. How it works:

  * It firstly make rugplot of the data.
  * It then decide the kernel function and chooses mostly Gaussian distribution. So, it then creates one Gaussian curve (bell shaped) on top of each tick of rugplot.
  * Combine those all cruves and make one curve called KDE.

4. In KDE, y-axis shows density of each datapoint means likelihood of that datapoint.

```python
sns.kdeplot(x = 'salary', data = df);
```

If we want to see the KDE of only datapoints between e.g. 80000 and 140000 then use **clip** parameter.

```python
sns.kdeplot(x = 'salary', data = df, clip = [80000, 120000]);
```

We can change the smoothness of KDE using changing bandwidth of the curve using **bw_adjust** parameter. Making it more small add more noise to the KDE and making it very big gives a very smooth curve and ignore many data points. So use it wisely if wanna use.

```python
# using very small bandwidth: It will consider too much noise in the data

sns.kdeplot(x = 'salary', data = df, bw_adjust = 0.01);
```

```python
# using medium bandwidth: It covers most of the datapoints

sns.kdeplot(x = 'salary', data = df, bw_adjust = 0.5);
```

```python
# using very big bandwidth: It ignores most of the datapoints. NOT RECOMMENDED!

sns.kdeplot(x = 'salary', data = df, bw_adjust = 2, ls = '--', linewidth = 2, color = 'red');
```

We can add shade under the curve using **fill** parameter.

```python
sns.kdeplot(x = 'salary', data = df, bw_adjust = 0.5, fill = True);
```

# **4. Displot:**

Displots are same as histplots. But the difference is displot can show more than plots in it e.g. histplot, kde and rugplot all in one plot. But histplot show only histogram.

Woah !

All other properties will remain the same as histplot.

```python
sns.displot(x = 'salary', data = df);
```

Background color is one set is set permanently (because we set that in histogram and now it is set for displot too.) To get rid of it, restart the kernel of the jupyter notebook or set the style again.

```python
sns.set(style = 'white')
sns.displot(x = 'salary', data = df, kde = True, rug = True, color = 'yellow', edgecolor = 'red', ls = '--');
```
