# **Introduction**

* Categorical plots in seaborn are used to visualize and compare datapoints across different categories, to understand patterns and relationships within categorical variables. 

* They are particularly useful when examining the relationship between categorical and continous variables.

* It basically displays statistical metric (e.g. mean or std) per category. For example, mean value per category, count per category.

* It is visualization of groupby() call.

* It has 2 main types:

  1. countplot()
  2. barplot()


# **1. Countplot():**

1. It counts the number of rows per category in a column. So, display it for the **categorical variables**, for continous this will show nothing worthy.

2. In it, y-axis always shows the **count per category** and it **must start from 0**. Suppose we change the y-limit as:

plt.ylim(90, 260)

it will be decieving. Always display the information in trustful manner. So, just let the seaborn decide the limit of y-axis according to the data. Don't decide it by yourself. 

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('/content/sample_data/dm_office_sales.csv')
df.head()
```

```python
sns.countplot(x = 'division', data = df);
```

Names on x-axis are overlapped and difficult to read. So lets rotate them to make readable.

```python
import matplotlib.pyplot as plt

sns.countplot(x = 'division', data = df)
plt.xticks(rotation=45);
```

```python
# Using it for continous feature makes no sense. You can see its result below :/

sns.countplot(x = 'salary', data = df);
```

Useless!

We can add hue parameter in it too and the palette parameter too.

```python
sns.countplot(x = 'division', hue = 'level of education', data = df, palette = 'Set2');
```

# **2. Barplot:**

* It allows us to display any statistical metric per category e.g. mean or std.

* It also allows us to choose any feature to use as the y-axis as compared to countplot where we can only display count on y-axis.

* It has one parameter as estimator that we use to show what we want to calculate for the category of x-axis. For example, if I say I want to see the mean salary of each level of education, then we will use estimator as np.mean

* It has the other parameter as 'errorbar' which shows uncertainty for the datapoints of that category e.g. sd (standard deviation)

```python
import numpy as np

sns.barplot(x = 'level of education', y = 'salary', data = df, estimator = np.mean, errorbar = 'sd')
plt.xticks(rotation=45);
```

**Analysis:**

The bar in this plot shows the estimator (mean) value and the lines at top shows the errorbar. So we can safely say that 'People having associate's degree averagely earn 90000 salary'. We can also see from this plot that 'Salary of the Master's degree has lowest variability in its data or the salary values are less deviated from each other for the Masters degree people'.

We can add hue in barplot too.

```python
sns.barplot(x = 'level of education', y = 'salary', data = df, hue = 'division')
plt.xticks(rotation = 45);
plt.legend(loc= (1.1, 0.5));
```
