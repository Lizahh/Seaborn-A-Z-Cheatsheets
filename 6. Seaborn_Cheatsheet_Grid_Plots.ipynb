# **1. Catplot:** Creating Subplots based on grids

* We know making subplots in matplotlib. We can do it in seaborn using catplot.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/sample_data/StudentsPerformance.csv')
df.head()
```

```python
sns.catplot(x = 'gender', y = 'math score', data = df);
```

It is normal plot. What's special in it?

We can add extra parameters in it. 

Change the type of it using **kind parameter**. Add third dimension using **col parameter**.

```python
sns.catplot(x = 'gender', y = 'math score', data = df, kind = 'box', col = 'lunch');
```

Let's change the type to 'violin'.

```python
sns.catplot(x = 'gender', y = 'math score', data = df, kind = 'violin', col = 'lunch');
```

Let's change the type to 'point'.

```python
sns.catplot(x = 'gender', y = 'math score', data = df, kind = 'point', col = 'lunch');
```

Let's change the type to 'bar'.

```python
sns.catplot(x = 'gender', y = 'math score', data = df, kind = 'bar', col = 'lunch')
```

We can see when we create catplot it is saved as facetgrid. We can directly create facetgrid instead of creating it with catplot.

```python
g = sns.FacetGrid(data = df, col = 'lunch', hue = 'gender')
g = g.map(plt.scatter, 'math score', 'reading score');
g.add_legend();
```

# **2. Pairgrid:**

It is best thing. You can make grid plot. And decide all the plots in it by yourself.But its not super common.

```python
g = sns.PairGrid(df);
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.histplot)
```

Add some hue to it.

```python
g = sns.PairGrid(df, hue = 'gender');
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.histplot)
```

Even after adding hue, it never shows legend by itself. So add it by youself.

```python
g = sns.PairGrid(df, hue = 'gender');
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.histplot)
g = g.add_legend()
```
