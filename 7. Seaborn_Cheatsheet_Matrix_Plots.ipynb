# **Introduction:**

* Matrix plots are visual equivalent of displaying a pivot table.

* It displays all the numeric values in a dataframe.

* Not every Dataframe is a valid choice for a matrix plot.

* There are two types of matrix plot:

  1. Heatmap
  2. Clustermap

# **1. Heatmap:**

Heatmaps are used to visualize the relationship and patterns between two dimensions of data by representing the values as colors in a grid-like format, allowing for quick identification of high and low values and providing a concise overview of data distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=  pd.read_csv('/content/sample_data/country_table.csv')
df.head()
```

It has one column as categorical. Heatmap will give error. It wants all columns to be numerical. So instead of deleting, we can make the Countries column as index.

```python
df = df.set_index('Countries')
df.head()
```

In this data it is difficult to tell which regions are similar to each other based on all these rates. We can use heatmap to view the similarity using colors matrix.

```python
sns.heatmap(df)
```

Life expectancy column is totally different from all the remaining data. So lets just drop it.

```python
df = df.drop('Life expectancy', axis = 1)
```

```python
sns.heatmap(df);
```

To add space between cells, add **linewidth parameter.**

```python
sns.heatmap(df, linewidth = 0.5);
```

To write the values in boxes instead of colors only to better get the idea, we will use **annot parameter.**

```python
sns.heatmap(df, linewidth = 0.5, annot = True);
```

# **2. Clustermap:**

The purpose of clustermap in seaborn is to create a hierarchical clustering heatmap that organizes and **groups similar data points together based on their similarities.**

It automatically cluster similar groupings.

```python
sns.clustermap(df);
```

The main purpose of it to show the association of clusters in our data.

So in it, similar country regions and similar rates are going to be grouped together in a hierarchy.

We can also show clusters of only rows as following:

```python
sns.clustermap(df, col_cluster=False);
```
