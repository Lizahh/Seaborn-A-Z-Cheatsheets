# **1. Jointplot():**

* Joinplot provides a combination of scatter plot, histograms, hexplot, and kernel density estimates (KDE) to display the joint distribution and individual distributions of the variables. 

* The basic jointplot is based on histogram and scatterplot.

* It needs 2 continous features to plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/sample_data/StudentsPerformance.csv')
df.head()
```

```python
sns.jointplot(x = 'math score', y = 'reading score', data = df);
```

We can change the kind of intenal plot using **'kind'** parameter. Let's do it with 'hex' to get hexplot (hexagons).

```python
sns.jointplot(x = 'math score', y = 'reading score', data = df, kind = 'hex');
```

Let's change the internal kind of plot to histogram. 

```python
sns.jointplot(x = 'math score', y = 'reading score', data = df, kind = 'hist');
```

Let's change it back to default (scatterplot) and add the alpha parameter in it to change its opacity/visibility.

```python
sns.jointplot(x = 'math score', y = 'reading score', data = df, kind = 'scatter', alpha = 0.2);
```

Let's change the kind to kde.

```python
sns.jointplot(x = 'math score', y = 'reading score', data = df, kind = 'kde');
```

Let's fill the kde with shade.

```python
sns.jointplot(x = 'math score', y = 'reading score', data = df, kind = 'kde', fill = True);
```

Let's add some hue and fill it but with some opacity.

```python
sns.jointplot(x = 'math score', y = 'reading score', data = df, kind = 'kde', fill = True, alpha = 0.2);
```

Let's change the color of palette.

```python
sns.jointplot(x = 'math score', y = 'reading score', data = df, kind = 'kde', fill = True, palette = 'Set5');
```

Let's add some hue.

```python
sns.jointplot(x='math score', y='reading score', data=df, kind='kde', fill=True, hue='race/ethnicity');
```

# **2. Pairplot:**

* Pairplot is a quick way to compare all the continous features in the dataset.

* It automatically creates histogram for each column and a scatterplot comparison between all possible combinations of columns.

* It is super common plot in machine learning.

* Warning: Pairplot can be CPU and RAM intensive for large dataframes with many continous columns.

```python
sns.pairplot(data = df);
```

We can see the three graph in top most diagonal is showing the same scatteplots as which are shown in the lower most diagonal. To remove this redundancy, use the **corner** parameter.

```python
sns.pairplot(data = df, corner= True);
```

Let's add some hue and change the kind of diagonal plot as kde.

```python
sns.pairplot(data = df, hue = 'gender', diag_kind = 'kde');
```
